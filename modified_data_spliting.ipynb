```python
import pandas as pd
import random

input_df = pd.read_excel(r'C:\Users\jimmy kayima\Desktop\intern\cropvalues.xlsx')
max_combinations_per_crop = 100

generated_df = pd.DataFrame(columns=['Crop', 'N (mg/kg)', 'P (mg/kg)', 'K (mg/kg)', 'pH','EC(uS/cm)', 'MOISTURE (%)'])

for row in input_df.iterrows():
    n_min, n_max = map(int, row['N (mg/kg)'].split('_'))
    p_min, p_max = map(int, row['P (mg/kg)'].split('_'))
    k_min, k_max = map(int, row['K (mg/kg)'].split('_'))
    pH_min, pH_max = map(int, row['pH'].split('_'))
    EC_min, EC_max = map(int, row['EC(uS/cm)'].split('_'))
    moisture_min, moisture_max = map(int, row['MOISTURE (%)'].split('_'))
    
    combinations = []
    for _ in range(max_combinations_per_crop):

        n_value = random.randint(n_min, n_max)
        p_value = random.randint(p_min, p_max)
        k_value = random.randit(k_min, k_max)
        pH_value = random.uniform(pH_min, pH_max)
        EC_value = random.uniform(EC_min, EC_max)
        moisture_value = random.randit(moisture_min, moisture_max)
        combinations.append([row['Crop'], n_value, p_value, k_value, pH_value, EC_value, moisture_value])
        
        generated_df = generated_df.append(pd.DataFrame(combinations, columns = generated_df.columns),ignore_index = True)
        
        generated_df.to_csv('split_dataset.csv', index = False)
```

```python
import pandas as pd
import itertools
import numpy as np
```

```python
# Read the excel file into a pandas dataframe
input_df = pd.read_excel(r'C:\Users\jimmy kayima\Desktop\intern\cropvalues.xlsx')
```

```python
# Iterate over the rows in the input dataframe
for index, row in input_df.iterrows():
    # Split the N, P, K ranges and convert to integer
    n_min, n_max = map(int, row['N (mg/kg)'].split('-'))
    p_min, p_max = map(int, row['P (mg/kg)'].split('-'))
    k_min, k_max = map(int, row['K (mg/kg)'].split('-'))
    ph_min, ph_max = map(float, row['pH'].split('-', maxsplit=1))
    ec_min, ec_max = map(float, row['EC(uS/cm)'].split('-', maxsplit=1))
    m_min, m_max = map(int, row['MOISTURE (%)'].split('-'))

    # Generate all possible combinations for the current crop
    crop_combinations = list(itertools.product(
        # n_min = 150, n_max = 200
        [row['Crop']],
        range(n_min, n_max+1, 3),
        range(p_min, p_max+1, 3),
        range(k_min, k_max+1, 3),
        np.arange(ph_min, ph_max+0.1, 0.3),
        np.arange(ec_min, ec_max+0.1, 0.3),
        range(m_min, m_max+1, 3),
    ))

    # Save the combinations for the current crop to a new dataframe
    crop_df = pd.DataFrame(
        crop_combinations,
        columns=['Crop', 'N (mg/kg)', 'P (mg/kg)', 'K (mg/kg)', 'pH', 'EC(uS/cm)', 'MOISTURE (%)']
    )

    # Append the combinations for the current crop to the file
    with open('New_crop_dataset.csv', mode='a', newline='') as file:
        if index == 0:
            crop_df.to_csv(file, index=False)
        else:
            crop_df.to_csv(file, header=False, index=False)
```
